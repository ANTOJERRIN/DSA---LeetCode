```python
class Solution(object):
    def mapWordWeights(self, words, weights):
        """
        :type words: List[str]
        :type weights: List[int]
        :rtype: str
        """
        result_chars = []
        
        for word in words:
            current_word_weight = 0
            
            # Step 1: Calculate the total weight of the current word
            for char in word:
                # Get 0-25 index for the character
                char_index = ord(char) - ord('a')
                current_word_weight += weights[char_index]
            
            # Step 2: Apply modulo 26
            mod_value = current_word_weight % 26
            
            # Step 3: Map to reverse alphabetical order
            # ord('z') is 122. Subtracting mod_value shifts us backwards.
            mapped_char = chr(ord('z') - mod_value)
            
            # Add to our list of results
            result_chars.append(mapped_char)
            
        # Join the list of characters into a single string and return
        return "".join(result_chars)
```
