```python
from collections import deque

class Solution(object):
    def processStr(self, s):
        """
        :type s: str
        :rtype: str
        """
        # 1. Bi-directional queue container for O(1) endpoints edits
        buffer = deque()
        
        # 2. Virtual orientation toggle flag
        is_reversed = False
        
        for char in s:
            if char == '*':
                if buffer:
                    if is_reversed:
                        buffer.popleft()  # End of string is on the left
                    else:
                        buffer.pop()      # End of string is on the right
                        
            elif char == '#':
                # No limits here. Let Python's optimized C-backend handle the memory allocation.
                if is_reversed:
                    # Duplicate to the left side
                    snapshot = list(buffer)
                    buffer.extendleft(reversed(snapshot))
                else:
                    # Duplicate to the right side
                    buffer.extend(list(buffer))
                    
            elif char == '%':
                # Flip orientation direction instantly
                is_reversed = not is_reversed
                
            else:
                # Regular lowercase character insertion
                if is_reversed:
                    buffer.appendleft(char)
                else:
                    buffer.append(char)
                    
        # 3. Final single-pass string construction
        result = "".join(buffer)
        
        # 4. Return correct final orientation variant
        return result[::-1] if is_reversed else result
```
