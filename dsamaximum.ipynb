```python
from collections import Counter

class Solution(object):
    def maxNumberOfBalloons(self, text):
        """
        :type text: str
        :rtype: int
        """
        # 1. Count the raw frequencies of all characters in the text
        counts = Counter(text)
        
        # 2. Divide by the occurrences needed to form a single "balloon"
        possible_b = counts['b'] // 1
        possible_a = counts['a'] // 1
        possible_l = counts['l'] // 2  # Requires 2 characters per word
        possible_o = counts['o'] // 2  # Requires 2 characters per word
        possible_n = counts['n'] // 1
        
        # 3. The bottleneck item determines our global limit bound
        return min(possible_b, possible_a, possible_l, possible_o, possible_n)
```
