```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def deleteMiddle(self, head):
        # Check if the list contains 0 or 1 nodes. 
        # If head.next is None, the list has 1 node. Removing it leaves it empty (None).
        if not head or not head.next:
            return None
        
        # Initialize the tracking pointer at the absolute start of the chain.
        slow = head
        
        # Advance the forward pointer by two full steps immediately.
        # This setup establishes our required index offset.
        fast = head.next.next
        
        # Iterate through the chain. The loop terminates as soon as 'fast' 
        # reaches the last node or crosses past the final boundary.
        while fast and fast.next:
            slow = slow.next          # Advance slow by exactly 1 position
            fast = fast.next.next     # Advance fast by exactly 2 positions
        
        # The loop has terminated. 'slow' rests precisely on the node before the middle.
        # Reroute its 'next' reference to skip over the middle node entirely.
        slow.next = slow.next.next
        
        # Return the original head reference, which now points to the modified chain.
        return head
```
