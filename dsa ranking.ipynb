```python
class Solution(object):
    def arrayRankTransform(self, arr):
        """
        :type arr: List[int]
        :rtype: List[int]
        """
        # 1. Get sorted unique elements and map them to their ranks
        rank_map = {}
        for rank, val in enumerate(sorted(set(arr)), 1):
            rank_map[val] = rank
            
        # 2. Replace each element in the original array with its rank
        return [rank_map[num] for num in arr]
```
