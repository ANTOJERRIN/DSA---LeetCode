```python
class Solution(object):
    def countCompleteComponents(self, n, edges):
        # Build adjacency list
        adj = [[] for _ in range(n)]
        for u, v in edges:
            adj[u].append(v)
            adj[v].append(u)
            
        visited = [False] * n
        complete_components = 0
        
        for i in range(n):
            if not visited[i]:
                # Start traversal for a new component
                component_nodes = []
                queue = [i]
                visited[i] = True
                
                while queue:
                    curr = queue.pop(0)
                    component_nodes.append(curr)
                    for neighbor in adj[curr]:
                        if not visited[neighbor]:
                            visited[neighbor] = True
                            queue.append(neighbor)
                
                # Check if this component is complete
                # Every node must have a degree equal to (total nodes in component - 1)
                num_nodes = len(component_nodes)
                is_complete = True
                for node in component_nodes:
                    if len(adj[node]) != num_nodes - 1:
                        is_complete = False
                        break
                        
                if is_complete:
                    complete_components += 1
                    
        return complete_components
```
