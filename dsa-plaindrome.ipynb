```python
#plaindrome number dsa qn easy tyoe:
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False

        o = x
        r = 0

        while x > 0:
            d = x % 10
            r = r * 10 + d
            x = x // 10

        return o == r
```
