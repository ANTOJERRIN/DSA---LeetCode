```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        st=[]
        pairs={
            ")":"(",
            "}":"{",
            "]":"["

        }
        for ch in s:
            if ch in pairs.values():
                st.append(ch)
            else:
                if not st or st[-1]!=pairs[ch]:
                    return False
                st.pop()
        return len(st)==0               
```
