```python
class Solution(object):
    def stringIndices(self, wordsContainer, wordsQuery):
        """
        :type wordsContainer: List[str]
        :type wordsQuery: List[str]
        :rtype: List[int]
        """

        trie = [{"children": {}, "best": -1}]

        def is_better(new_idx, old_idx):
            if old_idx == -1:
                return True

            new_len = len(wordsContainer[new_idx])
            old_len = len(wordsContainer[old_idx])

            if new_len < old_len:
                return True

            if new_len == old_len and new_idx < old_idx:
                return True

            return False

        def update_best(node_idx, word_idx):
            if is_better(word_idx, trie[node_idx]["best"]):
                trie[node_idx]["best"] = word_idx

        for idx, word in enumerate(wordsContainer):
            node = 0
            update_best(node, idx)

            for ch in reversed(word):
                children = trie[node]["children"]

                if ch not in children:
                    children[ch] = len(trie)
                    trie.append({"children": {}, "best": -1})

                node = children[ch]
                update_best(node, idx)

        ans = []

        for query in wordsQuery:
            node = 0

            for ch in reversed(query):
                children = trie[node]["children"]

                if ch not in children:
                    break

                node = children[ch]

            ans.append(trie[node]["best"])

        return ans
```
