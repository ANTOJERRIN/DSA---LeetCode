```python
class Solution(object):
    def canReach(self, s, minJump, maxJump):
        """
        :type s: str
        :type minJump: int
        :type maxJump: int
        :rtype: bool
        """
        n=len(s)
        dp=[False]*n
        dp[0]=True
        rc=0

        for i in range(1,n):
            if i-minJump>=0 and dp[i-minJump]:
                rc+=1
            if i-maxJump-1>=0 and dp[i-maxJump-1]:
                rc-=1    
            if s[i]=="0" and rc>0:
                dp[i]=True
        return dp[n-1]            
       

        
```
