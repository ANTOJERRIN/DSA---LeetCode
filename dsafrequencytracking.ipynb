```python
class Solution(object):
    def findThePrefixCommonArray(self, A, B):
        """
        :type A: List[int]
        :type B: List[int]
        :rtype: List[int]
        """
        n=len(A)
        C=[0]*n
        sc=[0]*(n+1)
        c=0
        for i in range(n):
            sc[A[i]]+=1
            if sc[A[i]]==2:
                c+=1
            sc[B[i]]+=1
            if sc[B[i]]==2:
                c+=1
            C[i]=c
        return C          
        
```
