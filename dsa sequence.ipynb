```python
class Solution(object):
    def sequentialDigits(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: List[int]
        """
        sample = "123456789"
        result = []
        
        # Determine the minimum and maximum possible digit lengths
        min_len = len(str(low))
        max_len = len(str(high))
        
        # Slide a window of length 'length' over the sample string
        for length in range(min_len, max_len + 1):
            for start in range(10 - length):
                # Extract the substring and convert it to an integer
                num = int(sample[start:start + length])
                
                # Check if it fits within our target range
                if low <= num <= high:
                    result.append(num)
                    
        return result
```
