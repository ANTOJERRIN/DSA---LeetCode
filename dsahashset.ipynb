```python
class Solution(object):
    def longestCommonPrefix(self, arr1, arr2):
        """
        :type arr1: List[int]
        :type type arr2: List[int]
        :rtype: int
        """
        prefixes_set = set()
        
        # Step 1: Store all possible prefixes of numbers in arr1
        for num in arr1:
            while num > 0:
                prefixes_set.add(num)
                num //= 10
        
        max_len = 0
        
        # Step 2: Find the longest common prefix using arr2
        for num in arr2:
            curr_len = len(str(num))
            
            while num > 0:
                # Optimization: If the current number's length can't beat our max_len, stop checking
                if curr_len <= max_len:
                    break
                
                # If this prefix exists in our arr1 set
                if num in prefixes_set:
                    max_len = max(max_len, curr_len)
                    break # Found the longest prefix for this specific number
                
                num //= 10
                curr_len -= 1
                
        return max_len
        
```
