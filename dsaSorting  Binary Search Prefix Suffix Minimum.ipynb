```python
class Solution(object):
    def earliestFinishTime(self, landStartTime, landDuration, waterStartTime, waterDuration):
        """
        :type landStartTime: List[int]
        :type landDuration: List[int]
        :type waterStartTime: List[int]
        :type waterDuration: List[int]
        :rtype: int
        """
        from bisect import bisect_right

        def build(startTime, duration):
            rides = sorted(zip(startTime, duration))
            starts = [s for s, d in rides]
            n = len(rides)

            prefix_min_duration = [0] * n
            best = float("inf")

            for i in range(n):
                best = min(best, rides[i][1])
                prefix_min_duration[i] = best

            suffix_min_finish = [float("inf")] * (n + 1)
            best = float("inf")

            for i in range(n - 1, -1, -1):
                s, d = rides[i]
                best = min(best, s + d)
                suffix_min_finish[i] = best

            return starts, prefix_min_duration, suffix_min_finish

        def query(t, starts, prefix_min_duration, suffix_min_finish):
            idx = bisect_right(starts, t) - 1
            ans = float("inf")

            if idx >= 0:
                ans = min(ans, t + prefix_min_duration[idx])

            ans = min(ans, suffix_min_finish[idx + 1])

            return ans

        land_data = build(landStartTime, landDuration)
        water_data = build(waterStartTime, waterDuration)

        ans = float("inf")

        for i in range(len(landStartTime)):
            land_finish = landStartTime[i] + landDuration[i]
            ans = min(ans, query(land_finish, *water_data))

        for j in range(len(waterStartTime)):
            water_finish = waterStartTime[j] + waterDuration[j]
            ans = min(ans, query(water_finish, *land_data))

        return ans
```
