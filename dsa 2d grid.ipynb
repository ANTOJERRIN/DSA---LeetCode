```python
class Solution(object):
    def shiftGrid(self, grid, k):
        """
        :type grid: List[List[int]]
        :type k: int
        :rtype: List[List[int]]
        """
        m = len(grid)
        n = len(grid[0])
        total_elements = m * n
        
        # Optimize k to avoid unnecessary full rotations
        k = k % total_elements
        
        # Initialize the result grid with the same dimensions
        result = [[0] * n for _ in range(m)]
        
        for i in range(m):
            for j in range(n):
                # Flatten the current 2D index into a 1D index
                old_1d = i * n + j
                
                # Calculate the new 1D position after shifting
                new_1d = (old_1d + k) % total_elements
                
                # Convert the new 1D index back to 2D coordinates
                new_r = new_1d // n
                new_c = new_1d % n
                
                result[new_r][new_c] = grid[i][j]
                
        return result
```
